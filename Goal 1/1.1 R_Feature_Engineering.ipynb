```r
library(magrittr)
library(data.table)
```

```r
lending = read.csv('./DataSet/lending_club_processed.csv', stringsAsFactors = FALSE)
lending$X <- NULL
```

```r
onehot_transformation = function(targetF, dirr='', number){
    
    # 1. work out the c
    
    target = targetF
    
    
    new_column = vector()
    for (i in 1:number){
        new_column = c(new_column, paste(dirr, i-1, sep='_'))
    }
    
    # 2. work out the list
    
    new_column = as.character(new_column)
    
    for (i in 1:number){
        onehot = list()
        for (j in 1:number){
            onehot = c(onehot, list(0))
        }
        onehot[i] = 1
        numb = as.numeric(unlist(target[,..dirr]))
        targetF <- setDT(targetF) %>%
          .[grepl(i-1, numb), c(new_column) := ..onehot]
    }
    
}
```

```r
lending_onehot <- data.table(lending)

onehot_transformation(lending_onehot, 'home_ownership', 4)
lending_onehot <- subset(lending_onehot, select = -c(home_ownership))
onehot_transformation(lending_onehot, 'verification_status', 3)
lending_onehot <- subset(lending_onehot, select = -c(verification_status))
onehot_transformation(lending_onehot, 'purpose', 12)
lending_onehot <- subset(lending_onehot, select = -c(purpose))
onehot_transformation(lending_onehot, 'addr_state', 49)
lending_onehot <- subset(lending_onehot, select = -c(addr_state))
onehot_transformation(lending_onehot, 'initial_list_status', 2)
lending_onehot <- subset(lending_onehot, select = -c(initial_list_status))
onehot_transformation(lending_onehot, 'application_type', 2)
lending_onehot <- subset(lending_onehot, select = -c(application_type))
onehot_transformation(lending_onehot, 'hardship_flag', 2)
lending_onehot <- subset(lending_onehot, select = -c(hardship_flag))
onehot_transformation(lending_onehot, 'debt_settlement_flag', 2)
lending_onehot <- subset(lending_onehot, select = -c(debt_settlement_flag))

lending_onehot <- data.table(lending_onehot)
```

```r
lending_onehot$funded_amnt <- NULL
lending_onehot$term <- NULL

target = lending_onehot$target
lending_onehot$target <- NULL
lending_onehot$target = target
lending_onehot
```
